```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load dataset to pandas dataframe
clothes = pd.read_csv('/content/drive/MyDrive/Fatima/Big Data Project/myntra_products_catalog.csv')
clothes.head()
```

```python
# column count and data type validation
clothes.info()
```

```python
# Validate columns for NULL values
clothes.isnull().sum()
```

```python
# Validate dataset for duplicates
clothes.duplicated().sum()
```

```python
# Describe numeric values
clothes.describe()
```

```python
# Gender Proportion
GenderCount = clothes.Gender.value_counts()
GenderCount
```

```python
# Gender Proportion pie chart
plt.style.use('ggplot')
fig1, ax1 = plt.subplots(figsize=(7,7), facecolor='white')
ax1.pie(GenderCount, labels=GenderCount.values, labeldistance=1.05,
        autopct='%1.1f%%', textprops={'fontsize': 14})
plt.legend(labels=GenderCount.index, loc='upper left', fontsize=11)
plt.title('Proportion of Gender', fontsize=16)
plt.show()
```

```python
# Color Proportion Top 10
ColorCount = clothes.PrimaryColor.value_counts().sort_values()
ColorCount.sort_values(ascending=False).head(10)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 7))
ColorCount.plot(kind='barh')
plt.tick_params(axis='x', labelsize=15)
plt.title('Proportion Of Primary Color', fontsize=20)

for index, value in enumerate(ColorCount):
    plt.text(value, index, str(value))

plt.show()
```
